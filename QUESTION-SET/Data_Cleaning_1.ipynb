```python
content = """<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>School Story</title>
</head>
<body>
    <h1>The First Day of School</h1>
    <p>It was a bright, sunny morning when <b>Emily</b> woke up to the sound of her alarm clock ringing loudly. She jumped out of bed, feeling a mix of excitement and nervousness. Today was her first day at the new school.</p>
    
    <p>As she hurriedly got dressed, she remembered her mother's advice, <i>"Just be yourself and you'll make friends in no time."</i> Emily smiled and finished getting ready. After a quick breakfast, she grabbed her backpack and headed out the door.</p>
    
    <p>The school building was massive, with red brick walls and tall windows. As she walked through the <u>main entrance</u>, she noticed the hallway was bustling with students. Some were chatting in groups, while others were hurrying to their classrooms.</p>
    
    <p>Emily's first class was <a href="classrooms.html#history">History</a>. She found the room and took a seat near the middle. Just as she was settling in, a girl with curly hair and glasses sat down next to her.</p>
    
    <p>"Hi, I'm <em>Sophie</em>," she said with a friendly smile. "Are you new here?"</p>
    
    <p>"Yes, I'm Emily," she replied, feeling relieved to have someone to talk to. "I just moved here last week."</p>
    
    <p>Sophie nodded, "Well, welcome! If you need any help, just let me know."</p>
    
    <p>The teacher, Mr. <strong>Johnson</strong>, entered the room and began the lesson. He was an older man with a kind face and a gentle voice. Throughout the class, he made sure to engage the students with interesting stories about historical events.</p>
    
    <p>At lunchtime, Emily and Sophie decided to sit together. They found a table in the cafeteria and were soon joined by two other girls, <abbr title="Abigail">Abby</abbr> and <abbr title="Isabella">Izzy</abbr>.</p>
    
    <p>"So, how do you like the school so far?" Abby asked, opening her lunchbox.</p>
    
    <p>"It's really nice. Everyone seems friendly," Emily replied, taking a bite of her sandwich.</p>
    
    <p>Izzy nodded, "Yeah, it's a great place. We have a lot of fun activities too. Are you planning to join any clubs or sports teams?"</p>
    
    <p>Emily thought for a moment. "I'm not sure yet. I was thinking about joining the <abbr title="Photography Club">photo club</abbr> or maybe trying out for the soccer team."</p>
    
    <p>"Both are awesome choices," Sophie said. "I'm in the photo club, and Abby and Izzy are on the soccer team. You should definitely come to the meetings and practices!"</p>
    
    <p>Emily smiled, feeling more at ease. By the end of the day, she had attended all her classes and even made a few more friends. As she walked home, she felt happy and excited about the days to come.</p>
    
    <p>That night, she told her parents all about her first day. "It was great!" she exclaimed. "I think I'm going to like it here."</p>
    
    <p>Her mother hugged her, "I'm so glad to hear that, sweetheart. Remember, every new beginning is a chance to make wonderful memories."</p>
    
    <p>Emily nodded, feeling grateful for her new friends and the exciting opportunities ahead. As she drifted off to sleep, she couldn't wait to see what tomorrow would bring.</p>
</body>
</html>"""

```

```python
tags_to_remove = ['b', 'i', 'u', 'em', 'strong', 'abbr', 'a']
start_index = 0 #start searching

while True:
    start_index = content.find('<p>', start_index)
    if start_index == -1:
        break
    
    end_index = content.find('</p>', start_index)
    if end_index == -1:
        break
    
    paragraph = content[start_index + 3:end_index] # Extract the paragraph content (excluding the <p> tags)
    
    #Loop for tags_to_remove
    for i in tags_to_remove:
        open_tag = f"<{i}"
        close_tag = f"</{i}>"
        
        # Remove these tags from the paragraph
        while open_tag in paragraph:
            tag_start = paragraph.find(open_tag)
            tag_end = paragraph.find('>', tag_start)
            if tag_end != -1:
                paragraph = paragraph[:tag_start] + paragraph[tag_end + 1:]
        paragraph = paragraph.replace(close_tag, "")
    
    # Print the cleaned paragraph
    print(paragraph.strip())
    print()  # Print an empty line for readability
    
    # Move the start index to search for the next paragraph
    start_index = end_index + 4
```
