### The *Python random module* is a built-in module that provides functionalities related to random number generation. 

**random.choice(sequence)** - Returns a random element from the given sequence

```python
import random
li1=[1,2,3,4,5,6,7]
print(random.choice(li1)) 
```

```python
items = ['Electronics', 'Clothing', 'Books', 'Food', 'Toys']
selected = random.choice(items)
print(selected)
```

**random.seed(a)** - Initializes the random number generator with a seed value

**random.random()** - Returns a random float between 0 and 1

```python
random.seed(5) 
print(random.random()) 
print(random.random())
#will be the same every time you run the code with a seed of 5, providing consistency in the generated random values.
```

```python
import random
random.seed(10)
selrandom =[random.random() for _ in range(3)]
print(selrandom)
```

**random.randint(a, b)** - Returns a random integer between a given range a and b i.e. inclusive i.e. a=start, b

```python
r1 = random.randint(5, 15)
print("Random number between 5 and 15 is",r1)
r2 = random.randint(-10, -2)
print("Random number between -10 and -2 is",r2)
```

**random.uniform(a, b)** - Returns a random float between a and b

```python
prices = [random.uniform(10, 50) for _ in range(5)]
print(prices)
```

**random.sample(items, k)** - Returns a k-length list of unique elements chosen from the items

```python
products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Camera']
selected = random.sample(products, 3)
print(selected)
```

**random.shuffle(x)** - Shuffles the sequence x in place

```python
ids = [101, 102, 103, 104, 105]
random.shuffle(ids)
print(ids)
```

**random.gauss(a,b)** - Returns a random float from a Gaussian distribution

```python
s = [random.gauss(170, 10)]
print(s)
```

**random.normalvariate(a,b)** - Another method for generating random floats from a Gaussian distribution

```python
s = [random.normalvariate(70, 5) for _ in range(2)]
print(s)
```

**random.choices(items, weights=None, k=n)** - Returns a k-sized list of elements chosen from the items, considering weights if provided

```python
import random

print(random.randrange(10, 20))

list1 = ['a', 'b', 'c', 'd', 'e']
print(random.choice(list1))

random.shuffle(list1)
print(list1)

print(random.random())
```
